# Transformación

```python
df['CodigoEstadoCivil'] = df['CodigoEstadoCivil'].fillna(-1)
```

```python
#dfPersona reemplazar los nulos por el -1
def FuncionEstado(cadena):
    resultado = 0
    if cadena=='SOLTERO':
        resultado = 1
    elif cadena in("CASADO",'CASADO SEPARACION BIENES'):
        resultado = 2
    elif cadena=='DIVORCIADO':
        resultado=3
    elif cadena=='UNION LIBRE':
        resultado = 4
    elif cadena=='VIUDO':
        resultado = 5
    elif cadena=='NO REGISTRA':
        resultado = 1
    
    return resultado
```

```python
dfEstadoCivil['CodigoEstadoCivilFUNCION']=dfEstadoCivil['EstadCivil'].apply(FuncionEstado)
```

```python
import locale
locale.setlocale(locale.LC_TIME,'es_ES.UTF-8')
```

```python
fecha_actual=min_fecha
d={
    'FechaINT':int(fecha_actual.strftime('%Y%m%d')),
    'DiaDescripcion':fecha_actual.strftime('%A'),
    'MesDescripcion':fecha_actual.strftime('%B'),
    'Anio':fecha_actual.year,
    'Mes':fecha_actual.month,
    'Dia':fecha_actual.day,
    'Fecha':fecha_actual
}
```

```python
arreglo_fecha=[]
#min_fecha
#max_fecha
fecha_actual=min_fecha
while fecha_actual<=max_fecha:
    d={
    'FechaINT':int(fecha_actual.strftime('%Y%m%d')),
    'DiaDescripcion':fecha_actual.strftime('%A'),
    'MesDescripcion':fecha_actual.strftime('%B'),
    'Anio':fecha_actual.year,
    'Mes':fecha_actual.month,
    'Dia':fecha_actual.day,
    'Fecha':fecha_actual
    }
    arreglo_fecha.append(d)
    fecha_actual=fecha_actual+timedelta(days = 1)#Dateadd Sql
```

```python
dfFecha=pd.DataFrame(arreglo_fecha)
 
```
